```python
# pruebas de parametro
w = 0.09
b = -3.6
```

```python
x = np.linspace(0,data['BMI'].max(),100)
y = 1/(1+np.exp(-(w*x+b)))
```

```python
data.plot.scatter(x='BMI',y='BPMeds')
plt.plot(x, y, color='black')
plt.ylim(0,data['BPMeds'].max()*1.1)
plt.scatter(x, y, color='#A9E2F3')
# plt.grid()
plt.xlabel('Indice de Masa Corporal')
plt.ylabel('BPMeds')
plt.show()
```
