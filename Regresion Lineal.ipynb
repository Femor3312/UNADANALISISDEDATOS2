```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import os
```

```python
lineal=pd.read_csv("C:\dataset\data.csv")
```

```python
lineal.plot.scatter(x="metro", y="precio")
plt.show()
```

```python
RegreLine = linear_model.LinearRegression()
```

```python
metros = lineal["metro"].values.reshape((-1,1))
```

```python
modelo = RegreLine.fit(metros, lineal["precio"])
```

```python
print("Interseccion(b)", modelo.intercept_)
print("Pendiente(m)", modelo.coef_)
```

```python
entrada= [[7.5],[10.0],[12.5],[15.0],[17.5],[20.0],[22.5]]
predicciones=modelo.predict(entrada)
print(predicciones)
```

```python
lineal.plot.scatter(x="metro", y="precio", label='Datos')
plt.scatter(entrada, predicciones, color='red')
plt.plot(entrada, predicciones, color='black', label='Linea de Regresion')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
